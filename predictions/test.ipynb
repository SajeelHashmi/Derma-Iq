```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load original image (BGR to RGB for matplotlib)
original = cv2.imread("received.jpg")
original = cv2.resize(original, (256, 256))  # Resize to 1024x1024
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
# print(original_rgb.shape)

for i in range(11):  # mask indices
    for j in range(1, 10):  # thresholds 0.1 to 0.9
        mask_path = f"mask_{i}/thresh0.{j}.png"
        
        if not os.path.exists(mask_path):
            print(f"Missing: {mask_path}")
            continue

        # Load mask
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        # Make a copy to overlay on
        overlay = original_rgb.copy()

        # Create a green layer
        green = [0, 255, 0]

        # Boolean mask where mask == 255
        overlay[np.where(mask == 255)] = green

        # Plot and save
        plt.figure(figsize=(10, 5))
        plt.suptitle(f"Mask {i} - Threshold 0.{j}", fontsize=16)

        plt.subplot(1, 2, 1)
        plt.imshow(original_rgb)
        plt.title("Original")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(overlay)
        plt.title("Mask Overlay (Green)")
        plt.axis("off")

        output_path = f"mask_{i}/overlay_thresh0.{j}.png"
        plt.savefig(output_path, bbox_inches='tight')
        print(f"Saved: {output_path}")

        plt.show()
```
